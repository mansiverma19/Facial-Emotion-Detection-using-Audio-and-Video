```python
import os
import datetime
import pyaudio
import time
import wave
```

```python
labels = ['Neutral','Disgust','Fear','Sadness', 'Anger', 'Happiness', 'Surprise']
```

```python
audio_path=os.path.join('FER','audio')
if not os.path.exists(audio_path):
        !mkdir {audio_path}
```

```python
for label in labels:
    l_path = os.path.join(audio_path, label)
    if not os.path.exists(l_path):
        !mkdir {l_path}
```

```python
#Giving Parameters
duration = 5
sample_rate=44100
channels=2 
chunk_size=1024 
audio_format=pyaudio.paInt16

name= input('What is your name?\n')
attempt= input('Attmept No.?\n')

for label in labels:
    print("audio for label {}".format(label))
    time.sleep(13)
    print("start now....")

    filename = os.path.join(audio_path,label, f"{label}-{name}-{attempt}.wav")

    print(f"Recording audio for {duration} seconds. Saving to {filename}")

    # Initialize PyAudio
    p = pyaudio.PyAudio()

    # Open stream
    stream = p.open(format=audio_format,
                    channels=channels,
                    rate=sample_rate,
                    input=True,
                    frames_per_buffer=chunk_size)

    frames = []

    # Record audio
    for i in range(0, int(sample_rate / chunk_size * duration)):
        data = stream.read(chunk_size)
        frames.append(data)

    # Stop the stream
    stream.stop_stream()
    stream.close()
    p.terminate()

    # Save the recorded audio to a WAV file
    wf = wave.open(filename, 'wb')
    wf.setnchannels(channels)
    wf.setsampwidth(p.get_sample_size(audio_format))
    wf.setframerate(sample_rate)
    wf.writeframes(b''.join(frames))
    wf.close()

    print(f"Recording saved to {filename}")
```

# Adding Ravdess 

```python
Ravdess_path=os.path.join('FER','Ravdess')
emotions={'01':'Neutral', '03':'Happiness','04':'Sadness','05':'Anger','06':'Fear','07':'Disgust','08':'Surprise'}
```

```python
for actor in os.listdir(Ravdess_path):
    val='Neutral'
    cnt=0
    print(actor)
    for audio_file in os.listdir(os.path.join(Ravdess_path,actor)):
        if audio_file.split('-')[2]=='02':
            continue
        emotion=emotions[audio_file.split('-')[2]]
        print(emotion)
        if val==emotion:
            cnt+=1
        else:
            val=emotion
            cnt=1
        new_nm=emotions[audio_file.split('-')[2]]+'-'+actor+'-'+str(cnt)+'.wav'
        old_path=os.path.join(Ravdess_path,actor,audio_file)
        new_path=os.path.join(Ravdess_path,actor,new_nm)
#         print(new_path)
        try:
            os.rename(old_path, new_path)
            print(f"Renamed: {old_path} -> {new_path}")
        except Exception as e:
            # Handle the exception 
            print(f"Error processing {file_path}: {e}")
```

# Copying in folder

```python
import shutil

for actor in os.listdir(Ravdess_path):
    print(actor)
    for audio_file in os.listdir(os.path.join(Ravdess_path,actor)):
        if audio_file.split('-')[2]=='02':
            continue
        emotion=audio_file.split('-')[0]
#         print(emotion)
        try:
            source_path = os.path.join(Ravdess_path,actor,audio_file)
            destination_path = os.path.join(audio_path, emotion)
            shutil.copy(source_path, destination_path)
            print(f"Copied: {source_path} -> {destination_path}")
        except Exception as e:
            print("not original")
```
